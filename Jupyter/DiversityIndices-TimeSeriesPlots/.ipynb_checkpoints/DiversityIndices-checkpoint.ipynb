```python
%matplotlib notebook

from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

#import DiversityIndicesBiblio
from DiversityIndicesBiblio import *
```

```python
def PlotRichness(VirusDatafile,BacteriaDatafile,t1,t2,titulo):
    
    dataVirus = open(VirusDatafile).readlines()[1:]
    dataBacteria = open(BacteriaDatafile).readlines()[1:]
   
    Tb = []
    Tv = []
    Rb = []    
    Rv = []
    
    # T = row[0], R= row[1], H= row[2], E= row[3]    
    
    for i in range(t1-1,t2):
        row=dataVirus[i].split("\t")
        Tv.append(float(row[0]))
        Rv.append(float(row[1]))      
    
    for i in range(t1-1,t2):
        row=dataBacteria[i].split("\t")
        Tb.append(float(row[0]))
        Rb.append(float(row[1]))
        
    plt.figure(dpi= 130)

    plt.plot(Tv, Rv, '-', label='Virus')
    plt.plot(Tb, Rb, '-', label='Bacteria')
    plt.title('Richness'+'\n'+titulo)
    plt.grid(True)
    plt.legend(loc='upper left')
    plt.xlabel('time $t$ (hr)')
    plt.ylabel('Richness  $S$')
    
    plt.show()

    
    
def PlotShannonIndex(VirusDatafile,BacteriaDatafile,t1,t2,titulo):
    
    dataVirus = open(VirusDatafile).readlines()[1:]
    dataBacteria = open(BacteriaDatafile).readlines()[1:]
   
    Tb = []
    Tv = []
    Hb = []    
    Hv = []
    
    for i in range(t1-1,t2):
        row=dataVirus[i].split("\t")
        Tv.append(float(row[0]))
        Hv.append(float(row[2]))      
    
    for i in range(t1-1,t2):
        row=dataBacteria[i].split("\t")
        Tb.append(float(row[0]))
        Hb.append(float(row[2]))
    
    plt.figure(dpi= 130)
    
    plt.plot(Tv, Hv, label='Virus')
    plt.plot(Tb, Hb, 'r', label='Bacteria')
    plt.grid(True)
    plt.legend(loc='upper left')
    plt.title('Shannon Index'+'\n'+titulo)
    plt.xlabel('time $t$ (hr)')
    plt.ylabel('ShannonIndex  $H$')
    
def PlotEvenness(VirusDatafile,BacteriaDatafile,t1,t2,titulo):
    
    dataVirus = open(VirusDatafile).readlines()[1:]
    dataBacteria = open(BacteriaDatafile).readlines()[1:]
    
    Tb = []
    Tv = []
    Eb = []    
    Ev = []
    
    for i in range(t1-1,t2):
        row=dataVirus[i].split("\t")
        Tv.append(float(row[0]))
        Ev.append(float(row[3]))      
    
    for i in range(t1-1,t2):
        row=dataBacteria[i].split("\t")
        Tb.append(float(row[0]))
        Eb.append(float(row[3]))
    
    plt.figure(dpi= 130)
    
    plt.plot(Tv, Ev, 'b', label='Virus')
    plt.plot(Tb, Eb, 'r', label='Bacteria')
    plt.grid(True)
    plt.title('Evenness'+'\n'+titulo)  
    plt.legend(loc='upper left')
    plt.xlabel('time $t$ (hr)')
    plt.ylabel('Evenness  $E_H$')
```
