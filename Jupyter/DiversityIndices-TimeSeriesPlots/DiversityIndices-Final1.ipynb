```python
%matplotlib notebook

from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

#import DiversityIndicesBiblio
from DiversityIndicesBiblio import *
```

```python
def PlotRichness(Tempo,Virus,Bacterias,t1,t2):
    
    T = []
    B = []    
    V = []
    
    for t in range(t1-1,t2):
        T.append(float(Tempo[t]))
        V.append(float(Virus[t]))
        B.append(float(Bacterias[t]))
    
    figure(dpi= 130)

    plot(T, V, '-', label='Virus')
    plot(T, B, '-', label='Bacteria')
    grid(True)
    legend(loc='upper left')
    title('Diversity Indices ')
    xlabel('time $t$ (hr)')
    ylabel('Richness  $S$')
    
def ShannonIndex(Tempo,Virus,Bacterias,t1,t2):
    
    T = []
    B = []    
    V = []
    
    for t in range(t1-1,t2):
        T.append(float(Tempo[t]))
        B.append(float(Bacterias[t]))
        V.append(float(Virus[t]))
    
    figure(dpi= 130)
    plot(T, V, label='Virus')
    plot(T, B, 'r', label='Bacteria')
    grid(True)
    legend(loc='upper left')
    xlabel('time $t$ (hr)')
    ylabel('ShannonIndex  $H$')
    
def PlotEvenness(Tempo,Virus,Bacterias,t1,t2):
    
    T = []
    B = []    
    V = []
    
    for t in range(t1-1,t2):
        T.append(float(Tempo[t]))
        V.append(float(Virus[t]))
        B.append(float(Bacterias[t]))
    
    figure(dpi= 130)
    
    plot(T, V, 'b', label='Virus')
    plot(T, B, 'r', label='Bacteria')
    grid(True)
    legend(loc='upper left')
    xlabel('time $t$ (hr)')
    ylabel('Evenness  $E_H$')    
```
